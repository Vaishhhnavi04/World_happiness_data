### Understanding the correlation of happiness score and other factors that impact it.

```python
corr_mat = df[['Ladder score', 'Logged GDP per capita', 'Social support',
                          'Healthy life expectancy', 'Freedom to make life choices',
                          'Generosity', 'Perceptions of corruption']].corr()
sns.heatmap(corr_mat, annot=True, cmap='coolwarm' )
plt.title('Correlation Matrix')
plt.show()
```

```python
#c = df.corr(numeric_only=True)
print(corr_mat["Ladder score"].sort_values(ascending=False))
```

### How various factors seem to affect the ladder score
* Ladder Score has a strong positive correlation with Logged GDP per capita (0.790), Healthy life expectancy (0.768) and Social support(0.757). This suggests that countries with higher GDP per capita tend to have higher happiness scores and better health and social support contribute positively to happiness.
* Perceptions of corruption has a negative correlation (-0.421) with the Ladder Score, which means that countries with lower levels of perceived corruption tend to be happier.
* there is an inverse relationship between generosity and happiness in the dataset.

## Relationship between happiness and various factors: 

```python
df_factors = df[['Logged GDP per capita', 'Social support',
                          'Healthy life expectancy', 'Freedom to make life choices',
                          'Generosity', 'Perceptions of corruption']]
```

```python
for col in df_factors:
    plt.figure(figsize=(6,6))
    sns.scatterplot(x=df['Ladder score'],y=df[col],hue=df["Regional indicator"]) 
    plt.title("Relation between Happiness score and "+col)
    plt.legend(loc=(1.05, 0.5) ,fontsize=10)
    
```

#### Above scatter plots describe the reltionship between Happiness score(ladder score) and various factors that contribute to it.
##### The observations that can be made from these are:
###### INCOME i.e GDP per capita
* From the scatter plot we can see that, happiness and gdp have a linear relation, we can observe that most western european and North america and ANZ regions that have a high GDP also have their happiness scores on the higher side of the scale.
* Some Sub-Saharan african countries despitehaving a gdp of above average can be observed to have lower happiness score but since this case is only a few we can understand them to be outliers.Outliers could represent countries with unique characteristics that impact their happiness levels.
* Similarly the country with the lowest happiness score belongs to the south asian region but doesnt have the lowest gdp out of all. We can perform further detailed analysis of these countries to be able to mark either as an outlier.
###### FREEDOM
* Positive Correlation: The scatter plot generally forms an upward-sloping pattern from left to right, it indicates a positive correlation. As Freedom to Make Life Choices increases, the Happiness Score tends to increase as well. The conclusion would be that greater personal freedom is associated with higher happiness.
* clusters or groups of data points could lead to insights about regional patterns.
###### CORRUPTION
* This scatter plot also helps us the perception of corruption is most countries is higher than the average range.
* As observed above, North american and AZN region with the least perception of corruption have the highest happiness scores
* Negative Correlation: The scatter plot generally forms a downward-sloping pattern from left to right, it indicates a negative correlation.Although here it is not a significat downward slope we can notice the pattern hence we can conclude that they are inversly related.

## Conclusions: 

 **GDP per Capita (Income)**: It is the most influential factor positively affecting happiness scores.Countries with higher income levels tend to have higher happiness scores.

 **Life Expectancy**: Healthy life expectancy is another strong positive influence on happiness. Countries with longer life expectancies tend to have happier populations.

 **Social Support**:Nations with stronger social support systems tend to have higher happiness scores indicating the importance of community and relationships.

**Freedom**: Countries where individuals have more personal freedom and autonomy tend to have higher happiness scores.

**Generosity**: AS seen earlier generosity has a relatively weak impact on happiness scores (close to zero), it is not a major contributor to happiness. It suggests that the willingness to give or share within a society may not be a primary driver of overall happiness.

**Perceptions of Corruption**: Perceptions of corruption have a significant negative impact on happiness, indicating that trust and transparency in institutions are essential for overall well-being.
